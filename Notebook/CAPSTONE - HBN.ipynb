# Working on covid df

```python
covid.head()
```

```python
covid_col_list = ['DATE','TOTAL_CASES','NEW_CASES','TOTAL_CONFIRMED']
covid2 = covid[covid_col_list]
covid2
```

```python
covid2 = covid2.rename(columns = {'DATE':'date','TOTAL_CASES':'total_cases','NEW_CASES':'new_cases','TOTAL_CONFIRMED':'total_confirmed'})
covid2.tail(10)
```

```python
#Creating month column
covid2['month'] = pd.to_datetime(covid2['date']).dt.date.astype(str).str[-5:-3].astype(str)
```

```python
covid2
```

```python
#Creating year column
covid2['year'] = pd.to_datetime(covid2['date']).dt.date.astype(str).str[:-6].astype(str)
covid2
```

```python
#Creating df for only 2020 info
covid20=covid2.loc[covid2.year == '2020']
covid20 = covid20.reset_index(drop=True)
covid20
```

```python
#Creating df for only 2021 covid info
covid21=covid2.loc[covid2.year == '2021']
covid21 = covid21.reset_index(drop = True)
covid21
```

```python
#Merging to add month_name column
covid20 = pd.merge(covid20, months, left_on = 'month', right_on = 'month', how = 'inner')
covid20
```

```python
#Merging to add month name column
covid21 = pd.merge(covid21, months, left_on = 'month', right_on = 'month', how = 'inner')
covid21
```

```python
#Creating df to get max covid cases for every month in 2021
max_cases21 = covid21.groupby('month_name').agg({'total_cases': [np.max]})
max_cases21.columns = max_cases21.columns.map(''.join)
max_cases21 = max_cases21.rename(columns = {'index': 'Type', 'total_casesamax': 'monthly_max'})
max_cases21.reset_index()
```

```python
#Creating df to get min number of covid cases per month in 2021
min_cases21 = covid21.groupby('month_name').agg({'total_cases': [np.min]})
min_cases21.columns = min_cases21.columns.map(''.join)
min_cases21 = min_cases21.rename(columns = {'index': 'Type', 'total_casesamin': 'monthly_min'})
min_cases21.reset_index()
```

```python
min_cases21
```

```python
#Merging df's of 2021 month min and max covid cases.
cov21month=pd.merge(max_cases21,min_cases21, on='month_name')
cov21month.reset_index()
```

```python
#Creating number of covid cases per month in 2021
cov21month['cases_per_month'] = cov21month['monthly_max'] - cov21month['monthly_min']
```

```python
cov21month=cov21month.reindex(["Jan", 'Feb', "Mar",'Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
cov21month
```

```python
#Graph of 2021 covid cases
ax = cov21month['cases_per_month'].plot(kind = 'bar',
                                               figsize = (15,5),
                                               title = 'Number of TN Covid-19 Cases Per Month in 2021',
                                               color = 'aqua',
                                               rot = 0)
ax.set_xlabel('Month')
ax.set_ylabel('Number of Cases');
```

```python
#Creating df to calculate max number of covid cases per month in 2020
max_cases20 = covid20.groupby('month_name').agg({'total_cases': [np.max]})
max_cases20.columns = max_cases20.columns.map(''.join)
max_cases20 = max_cases20.rename(columns = {'index': 'Type', 'total_casesamax': 'monthly_max'})
max_cases20.reset_index()
```

```python
#Creating df to calculate min number of covid cases for each month in 2020
min_cases20 = covid20.groupby('month_name').agg({'total_cases': [np.min]})
min_cases20.columns = min_cases20.columns.map(''.join)
min_cases20 = min_cases20.rename(columns = {'index': 'Type', 'total_casesamin': 'monthly_min'})
min_cases20.reset_index()
```
